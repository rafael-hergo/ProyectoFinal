### Objetivo general

### Objetivos específicos
* Graficar los datos obtenidos de una titulación ácido-base.
* Hacer un ajuste de curvas con polinomios de diferentes grados
* Obvservar que polinomio se ajusta mejor a la curva y graficar el error cuadrático
* Comparar el ajuste obtenido con un realizado en una hoja de cálculo.

### Modelos que representan el problema

Con ayudad de un potenciometro (medidor de pH), se realizó una titulación de cerveza. Como base se utilizó hidróxido de sodio (NaOH).

Se titularon 15[mL] de cerveza a la cuál se le añadieron de manera gradual 25[mL] de NaOH a 0.1M. Se registro una lectura del potencial de hudrógeno cada 0.5[mL] agregados de NaOH.

Se realizo una tabla donde se agregaron todos los datos de las mediciones de pH de la solución en función del volumen de NaOh añadido.

```python
import pandas as pd
file_name = 'Datos_Proyecto_SM.csv'
datos = pd.read_csv(file_name)
datos
```

### Solución del problema

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

x = datos.iloc[:,1].values
y = datos.iloc[:,2].values

def pH_fun(b, x, y, n):
    f=b@[x**i for i in range (n+1)]
    return np.sum((y-f)**2)

#Ajuste grado 1
b1 = [0,1]
r1 = opt.minimize(pH_fun, b1, args=(x,y,1))

#Ajuste grado 2
b2 = [0,1,1]
r2 = opt.minimize(pH_fun, b2, args=(x,y,2))

#Ajuste grado 3
b3 = [0,1,1,1]
r3 = opt.minimize(pH_fun, b3, args=(x,y,3))

#Ajuste grado 4
b4 = [0,1,1,1,1]
r4 = opt.minimize(pH_fun, b4, args=(x,y,4))

#Ajuste grado 5
b5 = [0,1,1,1,1,1]
r5 = opt.minimize(pH_fun, b5, args=(x,y,5))

#Ajuste grado 6
b6 = [0,1,1,1,1,1,1]
r6 = opt.minimize(pH_fun, b6, args=(x,y,6))

#Ajuste grado 7
b7 = [0,1,1,1,1,1,1,1]
r7 = opt.minimize(pH_fun, b7, args=(x,y,7))
```

#### Datos graficados de la titulación acido-base

```python
def f(b,x,n):
    return b@[x**i for i in range (n+1)]

plt.figure(figsize=(8,6))
plt.plot(x,y,'og',label='Datos')
plt.xlabel('Volumen de NaOH')
plt.ylabel('pH de la cerveza')
plt.legend(loc='best')
plt.grid()
```

#### Ajuste lineal

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,'og',ms=3,label='Datos')
plt.plot(x,f(r1.x,x,1),'b',label='Ajuste grado 1')
plt.grid()
plt.xlabel('Volumen de NaOH')
plt.ylabel('pH de la cerveza')
plt.legend(loc='best')
```

#### Ajuste cuadrático

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,'og',ms=3,label='Datos')
plt.plot(x,f(r2.x,x,2),'b',label='Ajuste grado 2')
plt.grid()
plt.xlabel('Volumen de NaOH')
plt.ylabel('pH de la cerveza')
plt.legend(loc='best')
```

#### Ajuste cúbico

* Los ajustes polinomiales de grado 6 y 7 no se ajustan de ninguna manera en la curva de titulación por lo que no se tomaron en cuenta para calcular el error cuadrático

#### Cálculo del error cuadrático

```python
plt.plot(1,r1.fun,'ob',label='Error de ajuste grado 1')
plt.plot(2,r2.fun,'or',label='Error de ajuste grado 2')
plt.plot(3,r3.fun,'ok',label='Error de ajuste grado 3')
plt.plot(4,r4.fun,'og',label='Error de ajuste grado 4')
plt.plot(5,r5.fun,'oy',label='Error de ajuste grado 5')
plt.legend(loc='best')
plt.grid()
r1.fun,r2.fun,r3.fun,r4.fun,r5.fun
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,'or',ms=3,label='Datos')
plt.plot(x,f(r1.x,x,1),'--b',label='Ajuste grado 1')
plt.plot(x,f(r2.x,x,2),':k',label='Ajuste grado 2')
plt.plot(x,f(r3.x,x,3),'--m',label='Ajuste grado 3')
plt.plot(x,f(r4.x,x,4),':g',label='Ajuste grado 4')
plt.plot(x,f(r5.x,x,5),'--p',label='Ajuste grado 5')
plt.grid()
plt.xlabel('Volumen de NaOH')
plt.ylabel('pH de la cerveza')
plt.legend(loc='best')
```

 * El polinomio que proporciona una mejor curva de ajuste para el conjunto de datos es el **polinomio de grado 5**.
